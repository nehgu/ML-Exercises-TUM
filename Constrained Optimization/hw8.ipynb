### Constrained Optimization of a Convex function (using a Lagrangian function)

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.legend_handler import HandlerLine2D
import matplotlib.patches as mpatches
from matplotlib.ticker import MultipleLocator
%matplotlib inline
```

### Plot feasible points and solution

```python
x = np.linspace(-1, 5, 500)
y1 = x*x + 1
y2 = x*x - 6*x + 8
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
f0, = ax.plot(x, y1, label=r'$f_0(x)$')
f1, = ax.plot(x, y2, label=r'$f_1(x)$')
sol, = ax.plot([2], [5], 'or', label='Solution')

ax.annotate(r'$p^* = 5$', xy=(2, 5), xytext=(2.05, 4.3), fontsize=13)
ax.xaxis.set_minor_locator(MultipleLocator(0.25))
ax.yaxis.set_minor_locator(MultipleLocator(1.25))

ax.grid(True, which='major')
ax.vlines(2, 0, 5, 'r', 'dotted', lw=1.5)
ax.axvline(x=0, color='k')

ax.hlines(0, 2, 4, '#ffd02f', 'solid', lw=3, label='Feasible Points')
ax.hlines(0, -1, 2, 'k', 'solid')
ax.hlines(0, 4, 5, 'k', 'solid')
plt.xlabel('x')

plt.legend(fontsize=16, loc=2, handler_map={sol: HandlerLine2D(numpoints=1)})
```

### Plot the Lagrangian Objective fucntion for different values of alpha

```python
x = np.linspace(-20, 20, 500)
y1 = x*x + 1
y2 = x*x - 6*x + 8
alphas = [0, 0.5, 1, 1.5, 2, 3, 4, 5, 8]
f, axes = plt.subplots(3, 3, figsize=(20, 20))
for i in range(3):
    for j in range(3):
        f0, = axes[i,j].plot(x, y1, label=r'$f_0(x)')
        f1, = axes[i,j].plot(x, y2, label=r'$f_1(x)')
        alpha = alphas[3 * i + j]
        L = x*x + 1 + alpha * (x*x - 6*x + 8)
        lagrangian, = axes[i,j].plot(x, L, label=r'$L(x, \alpha)$')
        
        axes[i,j].set_autoscaley_on(False)
        axes[i,j].set_xlim([-1, 5])
        axes[i,j].set_ylim([-5, 30])
        axes[i,j].xaxis.set_minor_locator(MultipleLocator(0.25))
        axes[i,j].yaxis.set_minor_locator(MultipleLocator(1.25))
        axes[i,j].grid(True, which='major')
        axes[i,j].axhline(y=0, color='k')
        axes[i,j].axvline(x=0, color='k')
        if 3*i + j != 0:
            axes[i,j].axvspan(2, 4, facecolor='#ffd02f', alpha=0.5, linestyle=None)
        axes[i,j].set_title(r'$\alpha = {}$'.format(alpha), fontsize=14)
llo = mpatches.Patch(color='#ffd02f', alpha=0.5, label='Lagrangian < Objective')

f.legend((f0, f1, lagrangian, llo), (r'$f_0(x)$', r'$f_1(x)$', r'$L(x, \alpha)$', 'Lagrangian < Objective', 'Lagrangian > Objective'), loc='lower center',bbox_to_anchor=(0, 0.07, 1, 0),
           bbox_transform=plt.gcf().transFigure, ncol=5, fontsize=16)
```
